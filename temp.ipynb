```python
import pandas as pd #for data cleaning, feature engineering and data analysis
import matplotlib.pyplot as plt #for data visualization
import seaborn as sns #for visualization->advance package than matplotlib
```

```python
#scatter plot needed for visualization
temp=pd.read_csv('temperature.csv')
temp.head()
```

```python
plt.scatter(temp['Temperature (°C)'],temp['Humidity (%)'],
            color="green",edgecolor="black"
)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs. Humidity")
```

```python
#skewness=>measure of dispersion 
# Histogram=> single value
sns.histplot(temp['Temperature (°C)'],kde=True) #temperature is right skewed in here
```

```python
sns.histplot(temp['Humidity (%)'],kde=True) #humidity is normally skewed in here
```
